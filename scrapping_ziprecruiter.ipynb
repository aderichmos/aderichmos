```python
def get_archive_link(roles):
  url = f'http://web.archive.org/cdx/search/cdx?url=https://www.ziprecruiter.com/Jobs/{roles}&from=20130101&to=20230215&output=json'
  urls = rq.get(url).text
  parse_url = json.loads(urls) #parses the JSON from urls.
  # print(parse_url)
  url_list = []
  for i in range(1,len(parse_url)):
    orig_url = parse_url[i][2]
    tstamp = parse_url[i][1]
    waylink = tstamp+'/'+orig_url
    url_list.append(waylink)
  # print(url_list)

  ## Compiles final url pattern.
  final_list = []
  for url in url_list:
    final_url = 'https://web.archive.org/web/'+url
    final_list.append(final_url)
    # print(final_list)

  return final_list
```

```python
#extract company
def extract_company(div): 
    company = div.find_all(name="span", attrs={"class":"company_name"})
    for span in company:
        return (span.text.strip())
    return 'NOT_FOUND'



# extract job location
def extract_location(div):
    for span in div.findAll('span', attrs={'class': 'company_location'}):
        return (span.text)
    return 'NOT_FOUND'


# extract job salary
def extract_salary(div): 
    for div in div.find_all(name='div', attrs={"class":"value"}):
        return (div.text.strip())
    return('NOT_FOUND')



# extract job type
def extract_employment_type(div): 
    for div in div.find_all(name='div', attrs={"class":"value"}):
        return (div.text.strip())
    return('NOT_FOUND')
    


 #extract link of job description 
def extract_link(div): 
      for a in div.find_all(name='a', attrs={'class':'job_link'}):
        return (a['href'])
      return('NOT_FOUND')   


# extract job title
def extract_job_title(div):
    for h2 in div.find_all(name='h2', attrs={"class":"title"}):
        return (h2.text.strip())
    return('NOT_FOUND')


# extract jd summary
def extract_summary(div): 
  for p in div.findAll('p', attrs={'class': 'job_snippet'}):
        return (p.text)
  return 'NOT_FOUND'
```

```python
# define dataframe columns
df = pd.DataFrame(columns = [ 'job_qry','job_title', 
                             'company_name', 'location', 'summary', 
                             'salary','employment_type','link'])
```
